단순 기능 구현만 보이기

```python
my_list=[]
my_list.append("사과")
my_list.append("바나나")
my_list.append("체리")

print(my_list)
# 2. 수정 (첫 번째 요소 변경)
my_list[0] = "오렌지"
print(my_list)

# 3. 삭제 (바나나 삭제)
my_list.remove("바나나")
print(my_list)
# 4. 조회 (전체 출력)
print("현재 리스트:", my_list)
print(my_list)
# 5. 검색 (체리가 있는지 확인)
search_item = "체리"
if search_item in my_list:
    print(f"'{search_item}'가 리스트에 존재합니다.")
else:
    print(f"'{search_item}'가 리스트에 없습니다.")
```

```python
my_tuple = (1,2,3)
print(my_tuple[0])
# 5. 검색 (체리가 있는지 확인)
search_item = 1
if search_item in my_tuple:
    print(f"'{search_item}'가 튜플에 존재합니다.")
else:
    print(f"'{search_item}'가 튜플에 없습니다.")
```

```python
# 🔹 기본 실습 3: 셋 활용 (추가, 삭제, 조회, 검색)
# 문제:
# 셋을 활용하여 다음 기능을 수행하는 프로그램을 작성하세요.

# 사용자가 입력한 값을 셋에 추가
# 특정 값을 삭제
# 셋의 모든 요소 조회
# 특정 값이 셋에 있는지 검색

# 단순 기능 구현만 보이기

my_set = {1,2,3,4,5}
print(f"{my_set}")
my_set.add("100")
print(f"{my_set}")
my_set.discard(3)
print(f"{my_set}")
# 5. 검색 (체리가 있는지 확인)
search_item = "100"
if search_item in my_set:
    print(f"'{search_item}'가 셋에 존재합니다.")
else:
    print(f"'{search_item}'가 셋에 없습니다.")
```

```python
# 🔹 응용 실습 2: 상품 재고 관리 (딕셔너리 활용)
# 문제:
# 딕셔너리를 활용하여 상품 재고를 관리하는 프로그램을 작성하세요.
#
# 새로운 상품 추가
# 특정 상품의 재고 변경
# 특정 상품 삭제
# 전체 상품 목록 조회
# 특정 상품 검색
my_dict = {"사과" : 100, "바나나" : 200, "딸기" : 300}
print(my_dict)
# 추가
my_dict["포도"] = 400
print(my_dict)
# 변경
my_dict["포도"] = 500
print(my_dict)
# 조회
print(my_dict["바나나"])
# 5. 검색 (체리가 있는지 확인)
search_item = "사과"
if search_item in my_dict:
    print(f"'{search_item}'가 my_dict에, 수량: {my_dict[search_item]} 존재합니다.")
else:
    print(f"'{search_item}'가 my_dict에 없습니다.")
```
